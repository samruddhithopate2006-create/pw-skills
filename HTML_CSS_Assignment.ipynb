## Question 1 : What are semantic HTML elements? Why is using them important for web development?

### Answer

Semantic HTML elements are HTML tags that clearly describe the meaning and purpose of the content they contain. The word semantic means meaningful. These elements help both browsers and developers understand the structure of a webpage without reading all the content.

Before HTML5, developers mostly used generic tags like `<div>` and `<span>` to create webpages. Although these tags are useful, they do not describe what kind of content they contain. HTML5 introduced semantic elements to make webpages more meaningful, organized, and easier to understand.

**Common Semantic HTML Elements**

- `<header>`	Represents the top section of a webpage or section
- `<nav>`	Contains navigation links
- `<main>`	Contains the main content of the webpage
- `<section>`	Groups related content
- `<article>`	Represents an independent article or blog post
- `<aside>`	Contains side information like advertisements or sidebars
- `<footer>`	Represents the bottom section of the webpage
- `<figure>`	Contains images or diagrams
- `<details>`	Creates expandable content

**Example of Semantic HTML**

```html
<!DOCTYPE html>
<html>

<head>
    <title>Semantic HTML Example</title>
</head>

<body>

<header>
    <h1>Tech Blog</h1>
</header>

<nav>
    <a href="#">Home</a>
    <a href="#">About</a>
    <a href="#">Contact</a>
</nav>

<main>

<section>

<h2>Latest Articles</h2>

<article>

<h3>Introduction to HTML</h3>

<p>HTML is the basic language used to create web pages.</p>

</article>

</section>

<aside>

<h3>Popular Posts</h3>

<p>Learn CSS</p>

</aside>

</main>

<footer>

<p>Copyright © 2026</p>

</footer>

</body>

</html>
```
**Why Semantic HTML is Important**
-  Improves Readability

Semantic elements make HTML code easy to read and understand.

Instead of writing many `<div>` elements, meaningful tags explain the purpose of each section.

-  Better Search Engine Optimization (SEO)

Search engines like Google can understand webpage content more accurately.

This improves search rankings.

-  Improves Accessibility

Screen readers used by visually impaired users understand semantic elements much better.

This makes websites accessible to everyone.

-  Easier Maintenance

When developers return to a project after several months, semantic elements make the code easier to understand.

-  Better Browser Understanding

Browsers understand which content is navigation, footer, article, or main content.

-  Improves Code Organization

Each section of the webpage has a clearly defined purpose.

**Advantages of Semantic HTML**

- Easy to understand
- Improves SEO
- Better accessibility
- Cleaner code
- Easier debugging

## Question 3: How can you make an HTML form more accessible to users with disabilities?

### Answer

An accessible HTML form is a form that can be easily used by everyone, including people with disabilities such as visual impairment, hearing impairment, motor disabilities, and cognitive disabilities.

Accessible forms allow users to navigate using keyboards, screen readers, and other assistive technologies.

Creating accessible forms is an important part of modern web development because it ensures equal access for all users.

**Ways to Make an HTML Form Accessible**
-  Use Proper <label> Elements

Every input field should have a label so users know what information to enter.

Example:

```html
<label for="name">Full Name:</label>
<input type="text" id="name" name="name">
```

Labels help screen readers identify form controls correctly.

-  Associate Labels with Inputs

  Use the for attribute in the <label> tag and match it with the id of the input field.

This improves accessibility and allows users to click on the label to focus the input.

-  Use Appropriate Input Types

Choose the correct input type based on the data being collected.

Examples:

type="email"
type="password"
type="number"
type="date"
type="tel"

These provide better validation and user experience.

-  Use Placeholder Text Carefully

Placeholder text can provide hints, but it should not replace labels because placeholders disappear when users start typing.

- Group Related Fields Using `<fieldset>` and `<legend>`

These elements help organize related form controls and improve navigation for screen reader users.

Example:

```html
<fieldset>
<legend>Personal Information</legend>

<label for="name">Name</label>
<input type="text" id="name">

<label for="email">Email</label>
<input type="email" id="email">

</fieldset>
```
-  Mark Required Fields Clearly

Use the required attribute and indicate required fields in the label.

Example:

`<input type="email" required>`
-  Provide Clear Error Messages

Display meaningful error messages when the user enters incorrect information.

Example:

❌ Invalid input
✅ Please enter a valid email address.
-  Ensure Keyboard Accessibility

Users should be able to complete the entire form using only the keyboard.

Avoid controls that require a mouse.

-  Maintain Good Color Contrast

Text should have sufficient contrast with the background.

Do not rely only on color to convey information.

## Question 4: Identify and correct the errors in the following CSS code:

```css
p {
font-size: 16;
color: #333
margin-top 10px;
}
```

### Answer
Explanation of Errors
-  Missing Unit for font-size

The value 16 should include a unit such as px.

Incorrect:

font-size: 16;

Correct:

font-size: 16px;
-  Missing Semicolon After color

Each CSS property should end with a semicolon.

Incorrect:

color: #333

Correct:

color: #333;
-  Missing Colon in margin-top

A colon (:) must separate the property name and value.

Incorrect:

margin-top 10px;

Correct:

margin-top: 10px;
-  Proper Indentation

Indenting CSS properties improves readability and maintenance.
```css
p {
  font-size:16px;
  color:#333;
  margin-top:10px;
}
```

## Question 6: Explain the CSS box model and its components.

### Answer

The CSS Box Model is one of the most important concepts in web development. Every HTML element displayed on a webpage is considered a rectangular box. The box model explains how the size of an element is calculated and how space is created around it.

Whenever a browser displays an HTML element, it creates four different areas:
These four parts together form the CSS Box Model.
**Components of the Box Model**
- Content

The content is the actual information displayed inside the element.

Examples:

Text
Images
Videos
Tables

Example:

width:200px;

height:100px;
- Padding

Padding is the space between the content and the border.

It increases the inner spacing of an element.

Example

padding:20px;
- Border

Border surrounds the padding.

It separates the element from other elements.

Example

border:2px solid black;

Border properties include:
Border width,
Border style,
Border color
-  Margin

Margin is the outer space around an element.

It creates distance between neighboring elements.

Example

margin:20px;

**Complete Example**
```css
div{

width:250px;

height:120px;

padding:20px;

border:3px solid blue;

margin:30px;

background-color:lightyellow;

}
```

## Question 7: How do the relative, absolute, and fixed positioning properties differ in CSS?

### Answer

The position property in CSS controls where an element appears on a webpage. It allows developers to move elements from their normal position and place them at specific locations.

Three commonly used positioning values are:

Each behaves differently.

-  Relative Positioning

Syntax

position: relative;

Relative positioning moves an element relative to its original position.

The element still occupies its original space in the webpage.

Example
```css
.box{

position:relative;

top:20px;

left:30px;

}
```
- Absolute Positioning

Syntax

position:absolute;

Absolute positioning removes the element from the normal document flow.

It is positioned relative to the nearest positioned parent.

Example

```css
.box{

position:absolute;

top:50px;

left:80px;

}
```

- Fixed Positioning

Syntax

position:fixed;

A fixed element stays in the same position even when the page is scrolled.

Example

```css
.box{

position:fixed;

bottom:20px;

right:20px;

}
```
**Characteristics**

- Fixed relative to browser window.
- Always visible during scrolling.
- Used for chat buttons, navigation bars, "Back to Top" buttons.
- Comparison Table
- Feature	Relative	Absolute	Fixed

**Advantages**

- Relative
- Easy positioning
- Maintains layout
- Absolute
- Flexible placement


## Question 8: Write a CSS rule to set a background image for a <div> with the class .banner, ensuring the image covers the entire area without repeating.

### Answer

CSS Code
```css
.banner{

background-image:url("background.jpg");

background-repeat:no-repeat;

background-size:cover;

background-position:center;

}
```
Complete HTML and CSS Example
```html
<!DOCTYPE html>
<html>

<head>

<title>Banner Example</title>

<style>

.banner{

width:100%;

height:400px;

background-image:url("background.jpg");

background-repeat:no-repeat;

background-size:cover;

background-position:center;

}

</style>

</head>

<body>

<div class="banner">

<h1>Welcome to My Website</h1>

</div>

</body>

</html>
```
**Explanation of Each Property**
-  background-image

Sets an image as the background.

Example

background-image:url("background.jpg");
-  background-repeat

Controls image repetition.

no-repeat

prevents the image from repeating.

-  background-size
cover

Makes the image cover the entire `<div>` without leaving empty space. The image may be cropped slightly to maintain its aspect ratio.

-  background-position
center

Places the image at the center of the element.

**Advantages**
- Attractive webpage design.
- Full-screen banner appearance.
- No repeated images.
- Responsive design.
- Better visual presentation.
- Professional website look.

**Applications**
- Website banners
- Landing pages
- Hero sections
- Login pages
- Portfolio websites
