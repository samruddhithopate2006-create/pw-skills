## Question 1
**Explain the basic structure of an HTML page. Why is it important to follow this structure?**

### Answer

The basic structure of an HTML page is the standard format used to create every webpage. It helps browsers understand how to display the webpage correctly. Every HTML document contains important elements like the doctype declaration, html tag, head section, and body section. Following this structure ensures that the webpage works properly across different browsers.

```html
<!DOCTYPE html>
<html>
<head>
  <title>My Web Page</title>
</head>
<body>
  <h1>Welcome</h1>
  <p>This is my first HTML page.</p>
</body>
</html>
```
Explanation of Each Tag

1. `<!DOCTYPE html>`

Declares that the document is written in HTML5.
It tells the browser which version of HTML is being used.

2. `<html>`

It is the root element of an HTML document.
All HTML elements are written inside this tag.

3. `<head>`

Contains information about the webpage.
Includes the title, meta tags, CSS links, JavaScript files, etc.
The content inside the head section is not visible on the webpage.

4. `<title>`

Defines the title of the webpage.
The title appears on the browser tab.

5. `<body>`

Contains all the visible content of the webpage.
Headings, paragraphs, images, links, tables, forms, videos, and other elements are placed inside the body tag.
Importance of Following HTML Structure
It helps browsers display webpages correctly.
Makes the code organized and easy to understand.
Improves readability for developers.
Makes debugging easier.
Improves browser compatibility.
Helps search engines understand webpage content.
Supports accessibility tools such as screen readers.
Makes website maintenance easier.
Follows HTML standards.
Creates professional and error-free webpages.

**Importance**

## Question 3
**Look at the code below. Can you find and fix the mistake?**

```html
<!-- This is an HTML comment
<h1>Welcome to my website</h1>
```

### Answer

```html
<!-- This is an HTML comment -->

<h1>Welcome to my website</h1>
```

**Explanation:**

The mistake is that the HTML comment is not closed.

HTML comments always start with

`<!--`

and end with

`-->`

Since the closing comment tag was missing, the browser treated everything after it as a comment. As a result, the heading was not displayed. After adding the closing comment tag, the webpage works correctly.


## Question 6
**Ravi wants to leave a note in his HTML file so that only other developers
can see it, not the users on the website. How should he do this? Why are such notes useful?**

### Answer

Ravi should use HTML comments. HTML comments are special notes written inside the HTML code. These comments are visible only in the source code and are not displayed on the webpage when users visit the website. They are mainly used to explain the code or leave reminders for other developers.

**Syntax of HTML Comment**

    `<!--` This is an HTML comment `-->`
**Example**
```html
<!DOCTYPE html>
<html>

<head>
    <title>Comments Example</title>
</head>

<body>

<!-- Main Heading -->

<h1>Welcome to My Website</h1>

<!-- Navigation Bar Starts Here -->

<p>This website is created for learning HTML.</p>

<!-- Footer will be added later -->

</body>

</html>
```
**Why HTML Comments are Useful**
- HTML comments help explain the purpose of different sections of code.
- They improve the readability of HTML documents.
- They make the code easier to understand for other developers.
- Comments help developers remember important information.
- They can be used to temporarily disable HTML code during testing.
- Comments make debugging easier.
- They help organize large HTML files into sections.
- Comments improve teamwork when many developers work on the same project.
- They reduce confusion while maintaining old code.
- Comments do not affect the appearance or performance of the webpage because browsers ignore them.

**Advantages of HTML Comments**
- Easy to write.
- Hidden from website visitors.
- Improve documentation.
- Helpful during maintenance.
- Useful for future updates.
- Improve collaboration.
- Reduce coding mistakes.
- Make projects more professional.


## Question 7
**Here’s a form snippet. Can you spot the error and rewrite the correct
version?**

```html
<form>

<input type="email" name="email">

<select>
<option>India</option>
</select>

<button type="submit">Send</button>

</form>
```
##Answer

**Errors**
- The closing button tag is incorrect. It is written as `</buton>` instead of `</button>`.
- The email input should use type="email" so the browser validates email addresses.
- The submit button should use type="submit" to submit the form.
- Labels should be added to improve usability and accessibility.
- A placeholder can guide users on what to enter.
- The required attribute ensures the user enters a value before submitting.
- Proper indentation makes the code easier to read and maintain.
- Adding multiple country options makes the dropdown more useful.
- Wrapping the form in a complete HTML document makes it a valid standalone     webpage.
- The corrected code follows HTML5 standards and is easier to understand and maintain.

**Corrected Code **

```html
<!DOCTYPE html>
<html>

<head>
    <title>Correct Form</title>
</head>

<body>

<form>

    <label>Email:</label>

    <input
        type="email"
        name="email"
        placeholder="Enter your email"
        required>

    <br><br>

    <label>Country:</label>

    <select>

        <option>India</option>

        <option>USA</option>

        <option>Canada</option>

        <option>Australia</option>

    </select>

    <br><br>

    <button type="submit">
        Send
    </button>

</form>

</body>

</html>
```
