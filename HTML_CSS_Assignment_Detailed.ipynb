## Question 1 : What are semantic HTML elements? Why is using them important for web development?

### Answer

Semantic HTML elements describe the meaning of content. They include `<header>`, `<nav>`, `<main>`, `<section>`, `<article>`, `<aside>`, and `<footer>`.

Importance:
1. Better SEO.
2. Better accessibility.
3. Easier maintenance.
4. Readable code.
5. HTML5 standards.
6. Logical structure.
7. Team collaboration.
8. Browser compatibility.
9. Reduced `<div>` usage.
10. Professional websites.

```html
<header>Header</header>
<nav>Menu</nav>
<main><section><article>Blog</article></section></main>
<footer>Footer</footer>
```

## Question 2: You're designing a blog page. Which semantic elements would you use to structure the page, and why?

### Answer

Use `<header>`, `<nav>`, `<main>`, `<article>`, `<section>`, `<aside>`, and `<footer>` because they organize content, improve SEO, accessibility, readability, maintenance, browser understanding, teamwork, and standards compliance.

## Question 3: How can you make an HTML form more accessible to users with disabilities?

### Answer

1. Use labels.
2. Use `for` attribute.
3. Correct input types.
4. Required fields.
5. Fieldset and legend.
6. Keyboard accessibility.
7. Error messages.
8. Color contrast.
9. ARIA when needed.
10. Clear instructions.

```html
<form>
<label for="email">Email</label>
<input id="email" type="email" required>
<button type="submit">Submit</button>
</form>
```

## Question 6: Explain the CSS box model and its components.

### Answer

The CSS box model consists of Content, Padding, Border, and Margin. It controls spacing, layout, alignment, sizing, and responsive design.

```css
div{
 width:200px;
 padding:20px;
 border:2px solid black;
 margin:15px;
}
```

## Question 7: How do the relative, absolute, and fixed positioning properties differ in CSS?

### Answer

Relative: moves from original position. Absolute: positioned relative to nearest positioned ancestor and removed from normal flow. Fixed: stays fixed relative to viewport during scrolling.

```css
.relative{position:relative;top:20px;}
.absolute{position:absolute;top:20px;left:30px;}
.fixed{position:fixed;bottom:10px;right:10px;}
```

## Question 8: Write a CSS rule to set a background image for a <div> with the class .banner, ensuring the image covers the entire area without repeating.

### Answer

```css
.banner{
 background-image:url('background.jpg');
 background-repeat:no-repeat;
 background-size:cover;
 background-position:center;
}
```
